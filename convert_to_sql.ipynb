```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to SQL Server (Windows Authentication)
engine = create_engine(
    "mssql+pyodbc://localhost/olist_db"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

# Define files and table names
tables = {
    'customers':      r'data\olist_customers_dataset.csv',
    'orders':         r'data\olist_orders_dataset.csv',
    'order_items':    r'data\olist_order_items_dataset.csv',
    'order_payments': r'data\olist_order_payments_dataset.csv',
    'products':       r'data\olist_products_dataset.csv',
}

# Load each CSV into SQL Server
for table_name, path in tables.items():
    df = pd.read_csv(path, dtype=str)  # str keeps NULLs safe
    df.to_sql(
        name=table_name,
        con=engine,
        if_exists='replace',  # overwrites the empty tables
        index=False,
        chunksize=1000
    )
    print(f"{table_name}: {len(df):,} rows loaded")

print("\n All tables loaded successfully!")
```
