```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Connection
engine = create_engine(
    "mssql+pyodbc://@localhost/olist_db?driver=ODBC+Driver+17+for+SQL+Server"
)
```

```python
# Load tables
orders = pd.read_sql("SELECT * FROM dbo.orders", engine)
payments = pd.read_sql("SELECT * FROM dbo.order_payments", engine)
customers = pd.read_sql("SELECT * FROM dbo.customers", engine)
items = pd.read_sql("SELECT * FROM dbo.order_items", engine)
products = pd.read_sql("SELECT * FROM dbo.products", engine)
```

```python
# Merge all tables
df = orders.merge(payments, on="order_id") \
           .merge(customers, on="customer_id") \
           .merge(items, on="order_id") \
           .merge(products, on="product_id")
```

```python
# Clean data types
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["payment_value"] = pd.to_numeric(df["payment_value"], errors="coerce")
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
# Feature engineering
df["order_month"] = df["order_purchase_timestamp"].dt.to_period("M").astype(str)

# Export for BI
df.to_csv("../dashboard/olist_clean_dataset.csv", index=False)

print("Dataset ready for Power BI!")
```
